#### 6.- Realiza una función que realice una digestión con las siguientes enzimas de restricción en nuestra secuencia. La función tiene que devolver la siguiente información: (2 pto)
1. Enzimas que tienen sitios de corte en la secuencia y posición (o posiciones de corte)
2. Enzimas que no tienen sitios de corte (si las hay)
3. Resultado de la digestión: los fragmentos resultantes de nuestra secuencia
4. Indicar número y tamaño de cada fragmento

```python
#diccionario de enzimas
enzR = {
    'HindIII' : 'A*AGCTT',
    'BamHI' : 'G*GATCC',
    'AluI' : 'AG*CT',
    'Sau3AI' : '*GATC',
    'EcoRI' : 'G*AATTC'
}
def digestion(secuencia, enzR): #Recibe la secuencia de ADN y el diccionario de enzimas y va devolder un diccionario con todos los resultados
    secuencia = secuencia.upper() #convierte la secuencia a mayúsculas que es como lo hemos definido en el diccionario de enzimas de restricción
    enzimas_con_corte = {} #creamos un diccionario vacío para posteriormente guardar las enzimas que sí cortan. Usamos diccionario porque necesitamos guardar pares: enzima → lista de posiciones
    enzimas_sin_corte = [] #creamos una lista vacía para posteriormente guardar las enzimas que no cortan. Usamos lista porque solo necesitamos guardar nombres, no posiciones.

    #Búsqueda de sitios de corte 
    for enzima, patron in enzR.items():
        # El asterisco (*) indica el punto de corte
        partes = patron.split('*') #patron es dónde corta cada enzima. Con la función "split" se divide el string por donde encuentra un "*"
        antes, despues = partes[0], partes[1] #antes del "*" y después del "*"
        sitio = antes + despues #reconstruye el sitio de reconocimiento de la enzima antes de cortar

        posiciones = [] #Lista vacía donde guardaremos TODAS las posiciones de corte
        start = 0 # Para indicar que empezamos buscando desde el inicio de la secuencia
        while True:
            idx = secuencia.find(sitio, start) #"find()" busca el sitio a partir de la posición "start". Devuelve -1 si no encuentra nada y sí encuentra entonces devuelve la posición donde empieza el sitio.
            if idx == -1: #no encuentras mas sitios
                break #sale del bucle
            # posición real del corte (inicio + offset del *)
            corte = idx + len(antes) #"idx" nos da dónde empieza el sitio de corte y "len(antes)" nos dice cuántos caracteres hay antes del corte.
            posiciones.append(corte) #para guardar la posición de corte en la lista
            start = idx + 1 #Prepara la siguiente búsqueda desde la posición siguiente

        #Verificamos si la lista posiciones tiene elementos o esta vacía
        if posiciones: #Si la lista no está vacía
            enzimas_con_corte[enzima] = posiciones # Se agrega la enzima a la lista de  "cortadores". 
        else: ##Si la lista está vacía
            enzimas_sin_corte.append(enzima) # # Se agrega la enzima a la lista de  " no cortadores". 

    # Para generar fragmentos
    cortes_totales = sorted({pos for posiciones in enzimas_con_corte.values() for pos in posiciones}) # Aquí se realiza una comprensión de conjuntos para extraer TODAS las posiciones, convierte a set para eliminar duplicados y con"sorted()" ordena las posiciones.
    fragmentos = [] ## Lista vacía para guardar los fragmentos
    inicio = 0 ## Empezamos desde el inicio de la secuencia (posición 0)
    for corte in cortes_totales: #hacemos referencia a la lista ordenada con todas las posiciones donde las enzimas cortan.
        fragmentos.append(secuencia[inicio:corte]) #Esto corta la secuencia desde el índice inicio hasta la posición de corte.
        inicio = corte #Actualizamos inicio para preparar el siguiente fragmento, así el siguiente fragmento empieza justo donde terminó el fragmento anterior.

    # Mostrar resultados
    print(" Enzimas que cortan y sus posiciones:") #Mostrar enzimas que cortan y sus posiciones
    for enz, pos in enzimas_con_corte.items():
        print(f"  {enz}: {pos}")

    print("\n Enzimas que no cortan:") #Mostrar enzimas que no cortan. 
    if enzimas_sin_corte:
        print("  " + ", ".join(enzimas_sin_corte))
    else:
        print("  Ninguna")

    print("\n Fragmentos resultantes:") #Mostrar fragmentos resultantes con su tamaño.
    for i, frag in enumerate(fragmentos, 1):
        print(f"  Fragmento {i}: {frag} (tamaño: {len(frag)})")

    print(f"\nNúmero total de fragmentos: {len(fragmentos)}") #Mostrar el número total de fragmentos.

    return {
        "enzimas_con_corte": enzimas_con_corte,
        "enzimas_sin_corte": enzimas_sin_corte,
        "fragmentos": fragmentos
    }
```

#### 7.- Realiza una función que mediante la introducción de una secuencia de ADN devuelva lo siguiente: (3pto)
1. Todos los ORFs posibles que puedan surgir de la secuencia.
2. Selecciona los 6 ORFs más largos e indica si corresponden a la strand positiva o negativa.
3. Devuelve como mejor predicción, el más largo de todos.
4. Calcula el porcentaje de aminoácidos aromáticos de esa proteína elegida y su longitud relativa a la secuencia original de ADN.

```python
def analizar_orfs(secuencia):
    secuencia = secuencia.upper().replace(" ", "") #Convierte a mayúsculas y quita espacios.
    start_codon = "ATG" #Define codones de inicio.
    stop_codones = ["TAA", "TAG", "TGA"] #Define codones de stop.

    # Para obtener la hebra complementaria reversa ,coger la hebra original y revertir su orden (porque las direcciones son opuestas) y después complementar cada nucleótido (A→T, T→A, C→G, G→C).
    def reverse_complement(seq):
        complement = {"A":"T","T":"A","C":"G","G":"C"}
        return "".join(complement.get(b,b) for b in reversed(seq)) # "join" une todos los caracteres resultantes en una sola cadena.

    # Para traducir ORF a proteína
    def traducir(orf): #Convierte una secuencia de ADN (un ORF) en una secuencia de aminoácidos (proteína) usando la tabla estándar del código genético.
        codon_tabla = { # creamos un diccionario con la tabla estándar del codigo genetico donde cada tres nucleótidos corresponden a un aminoácido. 
            'TTT':'F','TTC':'F','TTA':'L','TTG':'L','CTT':'L','CTC':'L','CTA':'L','CTG':'L',
            'ATT':'I','ATC':'I','ATA':'I','ATG':'M','GTT':'V','GTC':'V','GTA':'V','GTG':'V',
            'TCT':'S','TCC':'S','TCA':'S','TCG':'S','CCT':'P','CCC':'P','CCA':'P','CCG':'P',
            'ACT':'T','ACC':'T','ACA':'T','ACG':'T','GCT':'A','GCC':'A','GCA':'A','GCG':'A',
            'TAT':'Y','TAC':'Y','TAA':'*','TAG':'*','CAT':'H','CAC':'H','CAA':'Q','CAG':'Q',
            'AAT':'N','AAC':'N','AAA':'K','AAG':'K','GAT':'D','GAC':'D','GAA':'E','GAG':'E',
            'TGT':'C','TGC':'C','TGA':'*','TGG':'W','CGT':'R','CGC':'R','CGA':'R','CGG':'R',
            'AGT':'S','AGC':'S','AGA':'R','AGG':'R','GGT':'G','GGC':'G','GGA':'G','GGG':'G'
        }
        proteina = "" #Aquí se creará la secuencia traducida, añadiendo un aminoácido cada vez.
        for i in range(0,len(orf)-2,3): #Recorrer la secuencia de ORF en bloques de 3 nucleótidos.
            codon = orf[i:i+3] #extrae grupos de 3 letras
            aa = codon_tabla.get(codon,"") #Busca el aminoácido correspondiente en la tabla, si no hay devuelve "".
            if aa=="*" or aa=="": #tanto si existe aminoácido correspondiente como si no, la traducción se detiene.
                break
            proteina += aa #Va construyendo la proteína uno a uno.
        return proteina

    # Para encontrar ORFs en una hebra
    def encontrar_orfs_heb(seq,strand): #Recibe una secuencia y la etiqueta de hebra (+ o -)
        orfs = [] #Creamos lista donde guardamos los ORFs
        for frame in range(3): #Recorrer los 3 marcos de lectura, hay 3 marcoos de lectura dependiendo de la posición del codón desde donde empieces a leer.
            i = frame
            while i <= len(seq)-3: #para no leer bases incompletas
                codon = seq[i:i+3] #obtiene un codón
                if codon == start_codon: #si encuentra un codón de inicio podría considerarse ORF
                    # Buscar codon stop
                    for j in range(i+3,len(seq)-2,3):
                        codon_stop = seq[j:j+3]
                        if codon_stop in stop_codones:
                            orf = seq[i:j+3] #Construir el ORF
                            orfs.append({ #Guardar el ORF encontrado, se guarda en el diccionario la secuencia del ORF, dónde empieza y dónde acaba, la longitud y si es hebra + ó -.
                                "orf":orf,
                                "inicio":i,
                                "fin":j+3,
                                "longitud":len(orf),
                                "hebra":strand
                            })
                            i = j + 3  # saltar ORF encontrado para no solaparlo
                            break
                    else:
                        i += 3
                        continue #si no enceuntra stop continúa buscando otro ATG
                else:
                    i +=3
        return orfs

    # ORFs hebra +
    orfs_pos = encontrar_orfs_heb(secuencia,"+")
    # ORFs hebra -
    orfs_neg = encontrar_orfs_heb(reverse_complement(secuencia),"-")

    # Todos los ORFs posibles
    todos_orfs = orfs_pos + orfs_neg
    todos_orfs.sort(key=lambda x: x["longitud"], reverse=True)

    # 6 ORFs más largos
    top6 = todos_orfs[:6]
    mejor_orf = todos_orfs[0] if todos_orfs else None

    # Calcular proteína, % aromáticos y longitud relativa
    #Aromáticos: F (Fenilalanina), Y (Tirosina), W (Triptófano)
    if mejor_orf: #Si existe un ORF (lista no vacía).
        proteina = traducir(mejor_orf["orf"]) #traduce ORF a proteínas, la función "traducir" convierte ADN en aminoácidos.
        aromaticos = sum(proteina.count(a) for a in ["F","Y","W"]) #se suman todos los armáticos
        porcentaje_aromaticos = (aromaticos / len(proteina))*100 if len(proteina)>0 else 0 #Se usa el if len(proteina)>0 por si la proteína está vacía.
        longitud_relativa = (mejor_orf["longitud"]/len(secuencia))*100
    else: #Si no hay ORFs
        proteina = ""
        porcentaje_aromaticos = 0
        longitud_relativa = 0

    #Imprimir todos los ORFs
    print(" TODOS los ORFs posibles:")
    for i, o in enumerate(todos_orfs,1):
        print(f"  ORF {i}: hebra {o['hebra']}, inicio {o['inicio']}, fin {o['fin']}, longitud {o['longitud']} pb")
    
    #Imprimir solo los 6 ORFs más largos
    print("\n 6 ORFs más largos:")
    for i,o in enumerate(top6,1):
        print(f"  ORF {i}: hebra {o['hebra']}, longitud {o['longitud']} pb")

    #Imprimir el mejor ORF
    print("\n Mejor ORF (predicción):")
    if mejor_orf:
        print(f"  Hebra: {mejor_orf['hebra']}")
        print(f"  Longitud: {mejor_orf['longitud']} pb")
        print(f"  Proteína traducida: {proteina}")
        print(f"  % Aromáticos: {porcentaje_aromaticos:.2f}%")
        print(f"  Longitud relativa al ADN: {longitud_relativa:.2f}%")
    else:
        print("  No se encontraron ORFs válidos.")

    return {
        "todos_orfs": todos_orfs,
        "top6": top6,
        "mejor": mejor_orf,
        "proteina": proteina,
        "porcentaje_aromaticos": porcentaje_aromaticos,
        "longitud_relativa": longitud_relativa
    }





```
